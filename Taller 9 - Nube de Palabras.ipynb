##Cargue Librerías y Dataset de Poemas##

```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')
import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
data = pd.read_csv("C:/Users/jlo46/OneDrive/Documentos/ESPECIALIZACION/SEMESTRE 2/Fundamentos NLP/poems.csv")
data
```

##Cargue de Imagen## 

```python
book = PIL.Image.open("C:/Users/jlo46/OneDrive/Documentos/ESPECIALIZACION/SEMESTRE 2/Fundamentos NLP/libro.png")
mask = np.array(book)
```

##Cargue de Nube de Palabras en Imagen## 

```python
palabras = data.content.str.cat(sep=' ')
wordcloud = WordCloud(width=800, height=400,
                     max_font_size=150, max_words=250,
                     background_color='white', colormap='cool',
                     stopwords=stopwords,
                     mask=mask, contour_width=0.5, contour_color='purple').generate(palabras) # https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
wordcloud.to_file("libro.png")
plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
