```python
import re
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from sklearn.feature_extraction.text import CountVectorizer
```

# Punto 1: Pre-Procesamiento

- `[11 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[11 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números y sin palabras vacias

#### Datos en bruto

```python
dialogos = pd.read_csv('C:/Users/jlo46/OneDrive/Documentos/ESPECIALIZACION/SEMESTRE 2/Fundamentos NLP/TALLER 4/dialogos.csv', header = 0)
df = pd.DataFrame(dialogos)
df
```

```python
df['Pre-Procesado'] = df['Locución'].apply(lambda Locución: [palabra for palabra in re.sub(r"[\W\d_+]", " ", Locución.lower()).split() if palabra not in stopwords_sp])
dialogos = df
dialogos
```
