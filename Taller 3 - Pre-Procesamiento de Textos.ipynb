# Punto 1:

- `[4pts]` Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con `with open`
- `[4pts]` Convertir a minúsculas
- `[4pts]` Conservar sólo palabras
- `[4pts]` Tokenizar

### Lectura del Archivo

```python
import pandas as pd
import os
with open ("C:/Users/jlo46/OneDrive/Documentos/ESPECIALIZACION/SEMESTRE 2/Fundamentos NLP/TALLER 3/UNA_SOLITARIA_VOZ_HUMANA.txt", encoding= "utf-8") as archivo:
    texto = archivo.read()
print(texto)
```

### En Minúsculas

```python
import re
minusculas= texto.lower() 
print(minusculas)
```

### Solo Palabras

```python
palabras = re.sub(r"[\W\d_]+"," ", minusculas)
print(palabras)
```

### Tokenización

```python
tokenizado = palabras.split()
print(tokenizado)
```

# Punto 2:

- `[8pts]` Quitar palabras vacias
- `[9pts]` ¿Cuáles son las 10 palabras no vacias más usadas?

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

### Sin palabras vacías

```python
texto = [palabra for palabra in tokenizado if palabra not in stopwords_sp]
texto
```

### 10 palabras no vacias más usadas

```python
from collections import Counter
Counter(texto).most_common(10)
```

# Punto 3:

- `[8pts]` Stemming del documento
- `[9pts]` ¿Cuáles son las 10 raíces más usadas?

### Stemming 

```python
from nltk.stem.snowball import SnowballStemmer
spanishStemmer = SnowballStemmer("spanish")
texto_stem = []
for palabra in texto:
    raiz = spanishStemmer.stem(palabra)
    texto_stem.append(raiz)
    
texto_stem
```

### 10 raíces más usadas 

```python
Counter(texto_stem).most_common(10)
```
